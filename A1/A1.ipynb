```python
# Importacion de los datos

import json

archivo = open('personas.json', 'r')
archivostring = archivo.read()
datos = json.loads(archivostring)
archivo.close()
```

```python
# Codigo de consultas

class Conslutas:

    def __init__(self):
        print('Digite 1 para la consulta de Antepasados en comun')
        print('Digite 2 para la consulta de Descendiente mas lejano')
        print('Digite 3 para la consulta de Familia nuclear mas grande')
        print('Digite 4 para la consulta Personas mas jovenes')
        print('Digite 5 para la consulta Agregar persona')
        variable = int(input('Que consulta quieres realizar?: '))

        if variable == 1:
            rut111 = input('Ingrese el rut de la persona 1: ')
            rut222 = input('Ingrese el rut de la persona 2: ')
            Conslutas.antepasadosencomun(rut111, rut222)

        if variable == 2:
            rutingresado111 = input('Ingrese el rut de la persona: ')
            Conslutas.descendientemaslejano(rutingresado111)
        
        if variable == 3:
            Conslutas.familiamasgrande()

        if variable == 4:
            numerok = input('Ingrese el numero entero k: ')
            Conslutas.masjovenes(numerok)


    def antepasadosencomun(rut1, rut2):
        antepasados1 = {}
        antepasados1[rut1] = datos[rut1]
        antepasados2 = {}
        antepasados2[rut2] = datos[rut2]
        contador = 0
        listo = 0
        n = 1
        notienen = 0
        exito = 0
        inicio = 0
        comienzo1 = 0
        comienzo2 = 0
        jejeje = 0

        if antepasados1[rut1]['apellido'] != antepasados2[rut2]['apellido']:
            n = 0
            print('No tienen antepasados en comun')

        elif antepasados1[rut1]['rut padre'] == antepasados2[rut2]['rut padre']:
            padre = antepasados1[rut1]['rut padre']
            madre = antepasados1[rut1]['rut madre']
            n = 0
            holaaaa = 2
            jejeje = 1
            print('El/Los antepasado/s en comun es/son: ')
            print(f'{datos[padre]["nombre"]} {datos[padre]["apellido"]} de rut {padre}')
            print(f'{datos[madre]["nombre"]} {datos[madre]["apellido"]} de rut {madre}')
            print(f'La distancia de parentesco es de {holaaaa}')
            
        while n > 0:
            tienepapas1 = 0
            tienepapas2 = 0
            papas1 = {}
            papas2 = {}
            papito1 = {}
            papito2 = {}

            for rutantepasados1 in antepasados1:

                for rutantepasados2 in antepasados2:

                    if inicio > 0:

                        if antepasados1[rutantepasados1]['rut padre'] == rutantepasados2 or antepasados1[rutantepasados1]['rut madre'] == rutantepasados2:
                            
                            if exito == 0:
                                exito += 1
                                print('El/Los antepasado/s en comun es/son: ')
                            
                            contador += 1
                            listo += 1
                            print(f'{antepasados2[rutantepasados2]["nombre"]} {antepasados2[rutantepasados2]["apellido"]} de rut {rutantepasados2}')
                        
                        elif antepasados2[rutantepasados2]['rut padre'] == rutantepasados1 or antepasados2[rutantepasados2]['rut madre'] == rutantepasados1:
                            
                            if exito == 0:
                                exito += 1
                                print('El/Los antepasado/s en comun es/son: ')

                            contador += 1
                            listo += 1
                            print(f'{antepasados1[rutantepasados1]["nombre"]} {antepasados1[rutantepasados1]["apellido"]} de rut {rutantepasados1}')

            if listo == 0:

                for rutante1 in antepasados1:

                    for rut in datos:

                        if datos[rutante1]['rut padre'] == rut or datos[rutante1]['rut madre'] == rut:
                            inicio += 1
                            tienepapas1 += 1
                            papas1[rut] = datos[rut]
                            papito1[rut] = datos[rut]

                if comienzo1 == 0 and tienepapas1 > 0:
                    comienzo1 += 1
                    antepasados1 = {}

                else:
                    for rut in antepasados1:
                        papas1[rut] = antepasados1[rut]

                if tienepapas1 > 0:
                    contador += 1

                    for rutpapas1 in papas1:

                        for rutcito2 in antepasados2:

                            if inicio > 2:

                                if papas1[rutpapas1]['rut padre'] == rutcito2 or papas1[rutpapas1]['rut madre'] == rutcito2:
                                    
                                    if exito == 0:
                                        exito += 1
                                        print('El/Los antepasado/s en comun es/son: ')

                                    contador += 1
                                    listo += 1
                                    print(f'{antepasados2[rutcito2]["nombre"]} {antepasados2[rutcito2]["apellido"]} de rut {rutcito2}')
                
                                elif antepasados2[rutcito2]['rut padre'] == rutpapas1 or antepasados2[rutcito2]['rut madre'] == rutpapas1:
                                    
                                    if exito == 0:
                                        exito += 1
                                        print('El/Los antepasado/s en comun es/son: ')
                                    
                                    contador += 1
                                    listo += 1
                                    print(f'{papas1[rutpapas1]["nombre"]} {papas1[rutpapas1]["apellido"]} de rut {rutpapas1}')

                if listo == 0:
                    
                    for rutante2 in antepasados2:

                        for rut in datos:

                            if datos[rutante2]['rut padre'] == rut or datos[rutante2]['rut madre'] == rut:
                                inicio += 1
                                tienepapas2 += 1
                                papas2[rut] = datos[rut]
                                papito2[rut] = datos[rut]

                    if comienzo2 == 0 and tienepapas2 > 0:
                        comienzo2 += 1
                        antepasados2 = {}
                    
                    else:
                        for rut in antepasados2:
                            papas2[rut] = antepasados2[rut]

                    if tienepapas2 > 0:
                        contador += 1

            antepasados1 = papas1
            antepasados2 = papas2

            if listo > 0:
                n -= 1

            if tienepapas1 == 0 and tienepapas2 == 0 and listo == 0:
                n -= 1
                print('No tienen antepasados en comun')
                notienen = 1

        if listo >= 2:
            distancia = contador - listo + 1
        
        else:
            distancia = contador

        if notienen == 0 and jejeje == 0:
            print(f'La distancia de parentesco es de {distancia}')


    def descendientemaslejano(rutingresado):
        progenitores = {}
        progenitores[rutingresado] = datos[rutingresado]
        n = 1
        exito = 0

        while n > 0:
            n -= 1
            hijos = {}
            contador = 0

            for rutprogenitores in progenitores:

                for rut in datos:

                    if rutprogenitores == datos[rut]['rut padre'] or rutprogenitores == datos[rut]['rut madre']:
                        hijos[rut] = datos[rut]
                        contador += 1
                        exito += 1

            if contador > 0:
                progenitores = hijos
                n += 1

            elif exito == 0 and contador == 0:
                print(f'La persona de rut {rutingresado} no tiene descendientes')
            else:
                print('El/Los descendiente/s mas lejano/s es/son')

                for ruts in progenitores:
                    print(f'{progenitores[ruts]["nombre"]} {progenitores[ruts]["apellido"]} de rut {ruts}')


    def familiamasgrande():
        contador = 0
        familia = {}
        contadorfinal = 0
        nombres = []

        for rut in datos:
            hijos = {}
            contador = 0
            activacion = 0

            for rutcito in datos:

                if datos[rutcito]['rut padre'] == rut or datos[rutcito]['rut madre'] == rut:
                    hijos[rutcito] = datos[rutcito]
                    contador +=1

                    if activacion == 0:
                        activacion += 1
                        for rutardo in datos:
                            if datos[rutcito]['rut padre'] == rutardo or datos[rutcito]['rut madre'] == rutardo:
                                hijos[rutardo] = datos[rutardo]
                                contador += 1

            if contadorfinal < contador:
                contadorfinal = contador
                familia = hijos
                cantidad = 1

            elif contadorfinal == contador:
                cantidad += 1

        for rut in familia:
            if familia[rut]['nombre'] in nombres:
                nadaaa = 0
            else:
                nombres.append(familia[rut]['nombre'])

        print(f'La familia nuclear mas grande es de {contadorfinal} integrantes')
        print(f'Sus nombres son: ')

        for nombrito in nombres:
            print(nombrito)

        if cantidad >= 2:
            print(f'Hay un total de {cantidad} familias nucleares de {contadorfinal} integrantes')
    

    def masjovenes(k):
        jovenes = datos
        n = int(k)

        if n >= 1:
            print(f'Las {k} personas de mayor rut son: ')
        
        else:
            print('No existe')

        while n > 0:
            n -= 1
            rutactual = 0

            for rut in jovenes:
                largo = len(rut)
                rutcorto = rut[0:(largo-2)]
                rutlisto = int(rutcorto)
                
                if rutlisto >= rutactual:
                    rutactual = rutlisto
                    rutactualstr = rut
            
            print(f'{jovenes[rutactualstr]["nombre"]} {jovenes[rutactualstr]["apellido"]} de rut {rutactualstr}')

            del jovenes[rutactualstr]
```
